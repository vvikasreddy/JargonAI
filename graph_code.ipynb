```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r"C:\Users\lcruz\Documents\NLP\t_v_loss.csv"
loss_df = pd.read_csv(file_path)

# setting data
epochs = loss_df["Epoch"]
training_loss = loss_df["Training Loss"]
validation_loss = loss_df["Validation Loss"]

# setting up graph
plt.figure(figsize=(10, 6))
plt.plot(epochs, training_loss, label="Training Loss", marker="o")
plt.plot(epochs, validation_loss, label="Validation Loss", marker="o", linestyle="--")

# adding labels
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training and Validation Loss by Epoch")
plt.legend()
plt.grid(True)
plt.show()
```
